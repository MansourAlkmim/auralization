# Impulse response of a rectangular room

using the image source method, adapted from Jacques Cuenca 2017

1. define physical and geometrical properties
    * room dimension
    * absorption coefficient / reflection 
    * define source position
    * define receiver position
    * number of reflection on wall


```python
%matplotlib inline
import librosa
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import itertools
import librosa.display
import IPython.display as ipd
```

```python
# sound speed
c = 344.0

# absorption coeff
alpha = .1
# magnitude of reflection coeff
R = np.sqrt(1 - alpha)

# room dimension
height = 4
width = 7

fig1 = plt.figure()
ax1 = fig1.add_subplot(111)
ax1.add_patch(
    patches.Rectangle((0, 0), width, height, alpha=.4, label='room'))

# source position : why complex?
r0 = width / 3 + 1j * height / 4.5
ax1.plot(np.real(r0),np.imag(r0),'ro', ms=7, label='source')

# receiver position
r = r0 - 1 + 1j
ax1.plot(np.real(r),np.imag(r),'gD', ms=7, label='receiver')
ax1.legend(['source', 'receiver','room'])
```

```python
# open the .wav file
source, sr = librosa.load('Hu_m4.wav')
# source is the length of the audio array and sr is the sample rate
print('sample rate:', sr)
# plot wave plot
plt.figure(figsize=(14,5))
plt.subplot(1, 2, 1)
librosa.display.waveplot(source, sr=sr)

plt.subplot(1, 2, 2)
Source = librosa.stft(source)
Sourcedb = librosa.amplitude_to_db(abs(Source))
librosa.display.specshow(Sourcedb, sr=sr, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')

ipd.Audio('Hu_m4.wav')
```

### Image source method

2. define the maximum domain for image source consideration
3. store position of image sources

```python
# estimation of max distance from TR60 reverberation time

Nr = 1; # number of reflections on walls (actually 36)

delta = 10
dsmax = R**( 2 * Nr ) * np.abs( r - r0 ) * 10**( delta / 20 )

# reflection index
m = np.linspace(-Nr, Nr, 2*Nr+1)
n = m
x_coord, y_coord = np.meshgrid(m, n) # matrix with reflection indices


# image sources position (isp)
isp = [ (2 * x_coord * width - np.real(r0) + 1j * ( 2 * y_coord * height - np.imag(r0))),
       (2 * x_coord * width - np.real(r0) + 1j * ( 2 * y_coord * height + np.imag(r0))),
       (2 * x_coord * width + np.real(r0) + 1j * ( 2 * y_coord * height + np.imag(r0))),
       (2 * x_coord * width + np.real(r0) + 1j * ( 2 * y_coord * height - np.imag(r0)))]

# Source amplitudes
As = [ R ** (np.abs(2 * x_coord) - np.sign( x_coord - 1/2)) * R ** (np.abs(2 * y_coord) - np.sign(y_coord - 1/2)),
       R ** (np.abs(2 * x_coord) - np.sign( x_coord - 1/2)) * R ** np.abs(2 * y_coord),
       R ** np.abs(2 * x_coord) * R ** np.abs(2 * y_coord),
       R ** np.abs(2 * x_coord) * R ** ( np.abs(2 * y_coord) - np.sign(y_coord - 1/2))]


# arrange all elements of isp in a single list
# concatenate the whole sequence of arrays along an axis:
isp_ = np.concatenate( isp, axis=0 )
# put them all in one list
isp__ = [j for i in isp_ for j in i]
# convert it to array
isp___ = np.array(isp__)

# source-receiver distances
ds = np.abs(isp___ - r)
# source-receiver angle (listener is looking towards +y)
dsa = np.angle(isp___ - r) - np.pi/2
dsa = np.mod(dsa, 2 * np.pi)

# sorting by increasing distances
ds = np.sort(ds)
# angles in same order as amplitude
index = np.argsort(ds)
dsa = dsa[index]

# propagation times from sources to receiver
ts = ds / c 

# arrange all elements of As in a single list
# concatenate the whole sequence of arrays along an axis:
As_ = np.concatenate( As, axis=0 )
# put them all in one list
As__ = [j for i in As_ for j in i]
# convert it to array
As___ = np.array(As__)

# source amplitudes in the same order as ds
As___  = As___[index]

# total simulation time of impulse response
T = ds[-1] / c 

# time axis
t = np.arange(0, (T - 1/sr), 1/sr)

# room impulse response initialize
impulse_response = np.zeros(len(t))

for index2, value in enumerate(ts):
    if ts[index2] <= T: # if propagation time not larger than total time
        # match = correspondence index of times ts(index) in time axis t
        match = np.argmin( np.abs( ts[index2] - t))
        # room impulse response
        impulse_response[match] = As___[index2] / (4*np.pi*ds[index2])
        
plt.figure(figsize=(7,5))
plt.subplot(1, 1, 1)
plt.plot(t, impulse_response, 'o-')
# Convolution with recording
receiver = np.convolve(impulse_response, source)
```

```python
# plot wave plot
plt.figure(figsize=(14,5))
plt.subplot(1, 2, 1)
librosa.display.waveplot(receiver, sr=sr)

plt.subplot(1, 2, 2)
Receiver = librosa.stft(receiver)
Receiverdb = librosa.amplitude_to_db(abs(Receiver))
librosa.display.specshow(Receiverdb, sr=sr, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')

# synthesized audio
ipd.Audio(np.real(Receiver), rate=sr)
```
